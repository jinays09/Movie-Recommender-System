```python
import ast
ast.literal_eval(df.iloc[0].genres)
#removed first single quotes so that we can access each dict with indices
```

```python
def genre_list(obj):
  l=[]
  obj=ast.literal_eval(obj)
  for i in obj:
    l.append(i['name'])
  return l
```

```python
df['genres']=df['genres'].apply(genre_list)
```

```python
def keyword_list(obj):
  l=[]
  obj=ast.literal_eval(obj)
  for i in obj:
    l.append(i['name'])
  return l
df['keywords']=df['keywords'].apply(keyword_list)
```

```python
df.iloc[0].keywords
```

```python
def cast_list(obj):
  l=[]
  obj=ast.literal_eval(obj)
  cnt=0
  for i in obj:
    if(cnt==3):
      break
    cnt+=1
    l.append(i['name'])
  return l
df['cast']=df['cast'].apply(cast_list)
```

```python
df.iloc[0].cast
```

```python
df['crew'][0]
#want to extract director
```

```python
def crew_director(obj):
  obj=ast.literal_eval(obj)
  l=[]
  for i in obj:
    if(i['job']=='Director' or i['job']=='director'):
      l.append(i['name'])
      break
  return l
```

```python
df['crew']=df['crew'].apply(crew_director)
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
  l=[]
  for i in text.split():
    l.append(ps.stem(i))
  return " ".join(l)
stem('ok yeah yes please pleases')
```

```python
df['tags']=df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
df
```

```python
vectors=cv.fit_transform(df['tags']).toarray()
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cs=cosine_similarity(vectors)
```

```python
cs
```

```python
cs[0].shape
```

```python
sorted(list(enumerate(cs[0])),reverse=True,key=lambda x:x[1])
```

```python
def recommend(movie):
  movie_ind=df[df['title']==movie].index[0]
  movies=sorted(list(enumerate(cs[movie_ind])),reverse=True,key=lambda x:x[1])[1:6]
  l=[]
  for i in range(5):
    l.append(df.iloc[movies[i][0]]['title'])
  return l
recommend('Avatar')
  
```
